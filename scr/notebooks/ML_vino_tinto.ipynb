Hago el mapeo antes de quitar las columnas para  poder comparar con el resultado que arroje después del test

```python
# Define la función de mapeo
def mapeo(valor):
    if valor <= 4:
        return 0
    elif valor == 5:
        return 1
    elif 6 <= valor <= 7:
        return 2
    elif 8 <= valor <= 9:
        return 3

# Aplica el mapeo y agrega la nueva columna 'category'
df_test['category'] = df_test['quality'].map(mapeo)
```
